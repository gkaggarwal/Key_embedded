```python
seq = open('sequence.dat', 'r')
seq = seq.readlines()
print(len(seq), len(seq[0]))

key = open('key.dat', 'r')
key = key.readlines()
print(len(key), key)

temp = open('template.dat', 'r')
temp = temp.readlines()
print(len(temp), temp)
```

```python
#Pre - processing
for i in range(len(seq)):
  seq[i] = seq[i].strip()

for i in range(len(temp)):
  temp[i] = temp[i].strip()

for i in range(len(key)):
  key[i] = key[i].strip()
```

```python
print(len(temp), temp)
print(len(seq), seq[0])
print(len(key), key)
```

```python
import numpy as np
from PIL import Image as im
c=0
temp1= seq[0:3]
for i in temp1:
  l = list(map(int, list(i[:])))
  l = [x*255 for x in l]
  conv = np.reshape(l,(50,20))
  print(conv.shape)
  im.fromarray(conv, mode='L').save('pic'+str(c)+'.png')
  c= c+1

```

```python

final = open("final.txt", 'w')
final2 = open("final2.txt", 'w')
for t in temp:
  tl = list(t)
  for i in range(len(seq)):
    currseq = seq[i]
    for k in range(len(key)):
      currkey = key[k]
      listseq = list(currseq)
      length = len(tl)
      index =0
      keyindex=0
      while index <= len(listseq)-length+1:
        subseq = listseq[index:index+length]
        if subseq == tl:
          listseq.insert(index+length,currkey[keyindex])
          keyindex = keyindex+1
          if keyindex >= len(currkey):
            final.write("".join(listseq) + "\n")
            final2.write("Key "+str(k)+" Seq"+str(i)+" Temp"+t+"\n")
            break
        index = index+1
final2.close()
final.close()
```

```python
final = open("final.txt","r")
final = final.readlines()
final = [x[:-1] for x in final]
top1 = open("dict1.txt", "w")
top2 = open("dict2.txt", "w")
top3 = open("dict3.txt", "w")
top4 = open("dict4.txt", "w")
topcumu1 = open("dict_cumu_1.txt", "w")
topcumu2 = open("dict_cumu_2.txt", "w")
topcumu3 = open("dict_cumu_3.txt", "w")
topcumu4 = open("dict_cumu_4.txt", "w")
di = dict()
dict_file = open("dict.txt","w")
for s in final:
  di2 =dict()
  for i in range(3, 8):
    temp_dict = dict()
    for id in range(0, len(s)-i+1):
      prob = s[id:id+i] 
      if prob in temp_dict:
        temp_dict[prob] = temp_dict[prob]+1
      else:
        temp_dict[prob] = 1
      if prob in di:
        di[prob] = di[prob]+1
      else:
        di[prob] = 1
      if prob in di2:
        di2[prob] = di2[prob]+1
      else:
        di2[prob] = 1
    temp_dict = sorted(temp_dict.items(), key=lambda x: x[1], reverse=True) 
    temp_list = list(temp_dict) 
    top1.write(temp_list[0][0] + ":" + str(temp_list[0][1]) +", ")
    top2.write(temp_list[0][0] + ":" + str(temp_list[0][1]) +", "+
               temp_list[1][0] + ":" + str(temp_list[1][1]) +", ")
    top3.write(temp_list[0][0] + ":" + str(temp_list[0][1]) +", "+
               temp_list[1][0] + ":" + str(temp_list[1][1]) +", " +
               temp_list[2][0] + ":" + str(temp_list[2][1]) +", ")
    top4.write(temp_list[0][0] + ":" + str(temp_list[0][1]) +", "+
               temp_list[1][0] + ":" + str(temp_list[1][1]) +", " +
               temp_list[2][0] + ":" + str(temp_list[2][1]) +", " +
               temp_list[3][0] + ":" + str(temp_list[3][1]) +", ")
  
  top1.write('\n')
  top2.write('\n')
  top3.write("\n")
  top4.write('\n')


  for key in di2:
    dict_file.write(key+":"+str(di2[key])+", ")
  dict_file.write("\n")

  
dict_file2 = open("dict_cumulative.txt","w")
di_list = list(di.items())
bit3  = sorted(di_list[0:8], key=lambda x: x[1], reverse=True) 
bit4 = sorted(di_list[8:25], key=lambda x: x[1], reverse=True) 
bit5 = sorted(di_list[25:57], key=lambda x: x[1], reverse=True) 
bit6 = sorted(di_list[57:121], key=lambda x: x[1], reverse=True) 
bit7 = sorted(di_list[121:249], key=lambda x: x[1], reverse=True) 

topcumu1.write(str(bit3[0]) +"\n" + str(bit4[0])+"\n" +str(bit5[0]) +"\n" + str(bit6[0]) + "\n" +str(bit7[0]))
topcumu2.write(str(bit3[0]) +"\n" + str(bit3[1]) +"\n" +
               str(bit4[0]) +"\n" + str(bit4[1]) +"\n" +
               str(bit5[0]) +"\n" + str(bit5[1]) +"\n" +
               str(bit6[0]) +"\n" + str(bit6[1]) +"\n" +
               str(bit7[0]) +"\n" + str(bit7[1])) 
topcumu3.write((str(bit3[0]) +"\n" + str(bit3[1]) +"\n" + str(bit3[2]) + "\n")+
               (str(bit4[0]) +"\n" + str(bit4[1]) +"\n" + str(bit4[2]) + "\n")+
               (str(bit5[0]) +"\n" + str(bit5[1]) +"\n" + str(bit5[2]) + "\n")+
               (str(bit6[0]) +"\n" + str(bit6[1]) +"\n" + str(bit6[2]) + "\n")+
               (str(bit7[0]) +"\n" + str(bit7[1]) +"\n" + str(bit7[2]) + "\n"))
topcumu4.write((str(bit3[0]) +"\n" + str(bit3[1]) +"\n" + str(bit3[2]) + "\n")+str(bit3[3]) + "\n"+
               (str(bit4[0]) +"\n" + str(bit4[1]) +"\n" + str(bit4[2]) + "\n")+str(bit4[3]) + "\n"+
               (str(bit5[0]) +"\n" + str(bit5[1]) +"\n" + str(bit5[2]) + "\n")+str(bit5[3]) + "\n"+
               (str(bit6[0]) +"\n" + str(bit6[1]) +"\n" + str(bit6[2]) + "\n")+str(bit6[3]) + "\n"+
               (str(bit7[0]) +"\n" + str(bit7[1]) +"\n" + str(bit7[2]) + "\n")+str(bit7[3]) + "\n")

dict_file.close()
dict_file2.close()
top1.close()
top2.close()
top3.close()
top4.close()
topcumu1.close()
topcumu2.close()
topcumu3.close()
topcumu4.close()
```
